###**Data Cleaning**##




```python
# Our first and foremost step for any dataset is performing data cleaning.
df_current.isnull().sum()
```

```python
# We can see that it will be a very vast method to drop or impute by selecting any individual column so let us consider a parramter of handling null values in those columns greater than 30 percent.
thirtypernull=df_current.isnull().sum()
thirtypernull=thirtypernull[thirtypernull.values>(0.30*len(thirtypernull))]
thirtypernull = list(thirtypernull[thirtypernull.values>=0.30].index)
df_current.drop(labels=thirtypernull,axis=1,inplace=True)
df_current.isnull().sum()
```

```python
# Let us analyze CNT_FAM_MEMBERS column
df_current.CNT_FAM_MEMBERS.isnull().sum()
```

```python
#We can perform some other task let us fill this column with median values
medianfilling=df_current['CNT_FAM_MEMBERS'].median()
df_current.fillna(medianfilling,inplace=True)
df_current.CNT_FAM_MEMBERS.isnull().sum()
```

```python
 # We also notice there are some irrelevant columns so let us remove them.
baseless = ['DAYS_REGISTRATION','FLAG_MOBIL','FLAG_EMP_PHONE','FLAG_WORK_PHONE','FLAG_CONT_MOBILE','FLAG_PHONE','FLAG_EMAIL','WEEKDAY_APPR_PROCESS_START','HOUR_APPR_PROCESS_START','LIVE_REGION_NOT_WORK_REGION',
'REG_CITY_NOT_LIVE_CITY','FLAG_DOCUMENT_18']
df_current=df_current.drop(labels=baseless,axis=1,inplace=True)
```

##Handling outliers

```python
sns.boxplot(df_current.AMT_CREDIT)
plt.show()
```

####By the boxplot and a vast difference between the values of 95th and 99th percentile we can clearly see outliers we can cap or bin to retain information while reducing impact.

```python
sns.boxplot(df_current.AMT_INCOME_TOTAL)
plt.show()
```

####By the boxplot we can see some outliers clearly above from the rest datapoints we can cap or delete the data if they distort the analysis or impute them with mean or median if they result from errors.

```python
#Let us remove them from 1.5 IQR rule
q1=df_current['AMT_INCOME_TOTAL'].quantile(0.25)
q3=df_current['AMT_INCOME_TOTAL'].quantile(0.75)
iqri=q3-q1
df_current=df_current.loc[(df_current['AMT_INCOME_TOTAL']>(q1-1.5*iqri))&(df_current['AMT_INCOME_TOTAL']<(q3+1.5*iqri))]
```

```python
#Same for 'AMT_CREDIT' column
q1=df_current['AMT_CREDIT'].quantile(0.25)
q3=df_current['AMT_CREDIT'].quantile(0.75)
iqri=q3-q1
df_current=df_current.loc[(df_current['AMT_CREDIT']>(q1-1.5*iqri))&(df_current['AMT_CREDIT']<(q3+1.5*iqri))]
```

```python
df_previous=pd.read_csv('/content/previous_application.csv')
df_previous.head()
```

```python
df_previous.shape
```

```python
df_previous.info()
```

**NOW SIMILARLY PERFORMING DATA CLEANING IN THE PREVIOUS CUSTOMER RECORDS CSV FILE BY SELECTING A PARAMETER OF 30 PERCENT**

```python
nullvalues=df_previous.isnull().sum()
nullvalues=nullvalues[nullvalues.values>(0.3*len(nullvalues))]
nullvalues = list(nullvalues[nullvalues.values>=0.3].index)
df_previous.drop(labels=nullvalues,axis=1,inplace=True)
df_previous.isnull().sum()
```

**NOW UPON MERGING BOTH THE CSV FILES ON THE CUSTOMER ID COLUMN AND ANALYSING THE DATASET BY CONSIDERING VARIOUS PARAMETERS**

```python
updated_df=pd.merge(left=df_current,right=df_previous,how='inner',on='SK_ID_CURR')
updated_df.head()
```

**NOW LET US PERFORM UNIVARIATE ANALYSIS ON PORTFOLIO AND CONTRACT COLUMN**

```python
sns.countplot(x='NAME_PORTFOLIO',data=updated_df)
```

###WE SEE THAT MOST OF THE CUSTOMERS USE POS FOLLOWED BY CASH AND CARDS IS MINIMUM IN THIS CASE.

```python
sns.countplot(x='NAME_CONTRACT_TYPE_x',data=updated_df)
```

###NOW WE SEE MAJORITY OF THE CUSTOMERS USE CASH LOANS AS COMPARED TO REVOLVING LOANS.

**BROAD COMPARISON OF CURRENT LOAN DEFAULTERS WITH RESPECT TO PREVIOUS CUSTOMERS AND THE CATEGORY OF THE CLIENT**

```python
updated_df.pivot_table(values='TARGET',index='NAME_CONTRACT_STATUS',columns='NAME_CLIENT_TYPE',aggfunc='mean').plot.bar(figsize=(8,5),rot=45)
plt.xlabel('Previous loan status')
plt.ylabel('Current defaulters')
```
